## Tugas

Buatlah sebuah model KMeans dengan ketentuan,

1. Gunakan data 'Mall_Customers.csv'
2. Tentukan fitur apa yang tepat untuk melakukan clustering (minimal 2)
3. Buatlah model KMeans dengan mempertimbangkan jumlah $k$ yang terbaik.

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
# load data from data/Mall_Customers.csv
import pandas as pd
df = pd.read_csv('data/Mall_Customers.csv')
df.head()
```

```python

X = df.iloc[:,1:].values
```

```python
X.shape
```

```python
# Encode male and female in gender column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])
X[:5]
```

```python

# nd array to dataframe
df = pd.DataFrame(X)
df.head()
```

```python
# plot dataframe into scatter plot 3D
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df[1], df[2], df[3], c='blue', s=60)
ax.view_init(30, 185)
plt.show()
```

```python
# KMeans clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit_predict(X)
```

```python
kmeans.cluster_centers_
kmeans.labels_
```

```python
# plot the cluster with centroid in 3D

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df[1], df[2], df[3], c=kmeans.labels_, s=60, )
ax.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,3], c='blue', s=200, alpha=0.5)
ax.view_init(30, 185)
plt.show()


```

```python
# print inertia
kmeans.inertia_
```

```python
# search best k
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit_predict(X)
    inertia.append(kmeans.inertia_)

# plot inertia
plt.plot(range(1, 11), inertia, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.show()
```

```python
# print inertia from inertia list
for i in range(1, 11):
    print('k =', i, 'inertia =', inertia[i-1])
```

```python
# using k means with 6 cluster
kmeans = KMeans(n_clusters=6)
kmeans.fit_predict(X)
```

```python
# plot the cluster with centroid in 3D

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df[1], df[2], df[3], c=kmeans.labels_, s=60, )
ax.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,3], c='blue', s=200, alpha=0.5)
ax.view_init(30, 185)
plt.show()
```
